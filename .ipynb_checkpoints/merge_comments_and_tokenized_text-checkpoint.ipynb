# Preprocess Filter
1. Load data (in chunks)
2. Select only one category
3. Select only root comments
4. Filter out NaN

Overall, reduce the amount of data.

```python
import pandas as pd
pd.__version__
```

```python
filename = '/mnt/data/datasets/newspapers/guardian/comments.csv'

comments = pd.read_csv(filename)
```

```python
comments
```

```python
text = pd.read_table(skip_blank_lines=False)
```

```python
comments = pd.concat([comments, text], axis=1)
del text # free memory
```

```python
comments.to_csv('/mnt/data/group07/johannes/proc_data/pol_comments.csv')
print(comments.shape)
```
