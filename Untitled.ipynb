```python
co = pd.read_csv('/data/tmip/only_pol/comments_cleaned_100.csv')
```

```python
co['rank'] = co.groupby('article_id')['timestamp'].rank(method='dense').astype(int)
```

```python
co[co['article_id'] == 168]
```

```python
# sums = co.groupby('article_id')['upvotes'].sum()
co['total_upvotes'] = co.groupby('article_id')['upvotes'].transform('sum')
co['total_comments'] = co.groupby('article_id')['upvotes'].transform('count')
```

```python
co = co[co['total_upvotes'] != 0]
```

```python
co['score'] = co.apply(lambda row: (row.upvotes / math.log(row.total_upvotes + 1)) + (row['rank'] / row.total_comments), axis=1)
```

```python
co['bin'] = pd.qcut(co['score'], q=3, labels=range(3))
```
