# Merge Comments and Tokenized Text
The tokenized version only exists as text file. So first we have to merge them together.

```python
import pandas as pd
pd.__version__
```

```python
filename = '/mnt/data/datasets/newspapers/guardian/comments.csv'

comments = pd.read_csv(filename)
```

```python
comments
```

```python
text = pd.read_table('/mnt/data/datasets/newspapers/guardian/sorted_comments_twokenized_lower', skip_blank_lines=False)
```

```python
text
```

```python
# after verifiying that the data is correct, drop text so we can free memory
comments = comments.drop(['comment_text'], axis=1)
```

```python
comments
```

```python
comments = pd.concat([comments, text], axis=1)
del text # free memory
```

```python
comments
```

```python
comments = comments.rename(index=str, columns={"commen t_t ext": "comments_text"}) # the tokenized file has a strange col name
```

```python
comments = comments.rename(index=str, columns={"comments_text": "comment_text"}) # the tokenized file has a strange col name
```

```python
comments
```
