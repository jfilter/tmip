# get only root comments from the politics section

```python
import pandas as pd
pd.__version__
```

```python
# filename = '/data/tmip/raw/sorted_comments-standardized-tokenized.csv'
filename = '/mnt/data/datasets/newspapers/guardian/sorted_comments_twokenized_lower'

chunksize = 10 ** 6

comments_list = []

for chunk in pd.read_csv(filename, chunksize=chunksize):
    comment_chunk = chunk[chunk['comment_text'].notnull()] # filter out NaNs
    comment_chunk = comment_chunk[comment_chunk['parent_comment_id'].isnull()] # only select root comments
    comment_chunk = comment_chunk.drop(['parent_comment_id'], axis=1)
    print(comment_chunk.shape)
    comments_list.append(comment_chunk)
    print(len(comments_list))

# it's faster to first gather all in a list and concat once
comments = pd.concat(comments_list)
```

```python
%xdel comments_list
```

```python
comments.to_csv('/data/tmip/root_comments.csv')
```

```python

articles = pd.read_csv('/data/tmip/raw/articles-standardized.csv')
pol_a = articles[articles['article_url'].str.contains("https://www.theguardian.com/politics/")]
```

```python
# res = comments.set_index('article_id').join(pol_a.set_index('article_id'), how="inner").reset_index()
# res = res[res['article_url'].notnull()]
print(comments.shape)
comments = comments[comments['article_id'].isin(pol_a['article_id'])]
comments.to_csv('/data/tmip/only_pol/comments.csv')
print(comments.shape)
```
