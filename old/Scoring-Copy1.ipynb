```python
import pandas as pd
import math
import numpy as np
```

```python
# co = pd.read_csv('/data/tmip/root_comments.csv')
# co['rank'] = co.groupby('article_id')['timestamp'].rank(method='dense').astype(int)
# co = co[co['rank'] <= 10] # select only first 10 comments
# co = pd.read_csv('/data/tmip/root_comments_rank10.csv')
co = pd.read_csv('/data/tmip/root_comments_pol.csv')
co = co[co['rank'] <= 10] # select only first 10 comments
```

```python
co['total_upvotes'] = co.groupby('article_id')['upvotes'].transform('sum')
co['total_comments'] = co.groupby('article_id')['upvotes'].transform('count')
```

```python
co
```

```python
co = co[co['total_upvotes'] > 30] # do not consider articles with under 10 upvotes
```

```python
co = co[co['total_comments'] == 10] # remove articles with over 50 comments. 
```

```python
co
```

```python
co['rel_upvotes'] = co.apply(lambda row: (row.upvotes / row.total_upvotes) * 100, axis=1)
```

```python
filterd = co[co['comment_text'].str.len() > 100]
```

```python
filterd
```

```python
N = 1000
```

```python
res_pos = filterd.groupby("rank", group_keys=False).apply(lambda g: g.nlargest(N, "rel_upvotes"))
```

```python
res_pos
```

```python
res_neg = filterd.groupby("rank", group_keys=False).apply(lambda g: g.nsmallest(N, "rel_upvotes"))
```

```python
res_neg
```

```python
res_pos['bin'] = 1
res_neg['bin'] = 0
```

```python
res_pos
```

```python
res = pd.concat([res_pos, res_neg])
res.to_csv('/data/tmip/comments__pol_n1000.csv')
```
