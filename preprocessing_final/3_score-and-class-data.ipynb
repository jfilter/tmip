# Apply extracted features and filter

Uses
- /mnt/data/group07/johannes/proc_data/politics_comments.csv

For generating
- /mnt/data/group07/johannes/proc_data/classes_politics_comments_{top_bot_perc}.csv [top_bot_perc=1.0]



```python
import pandas as pd
import math
import numpy as np
```

```python
co = pd.read_csv('/mnt/data/group07/johannes/proc_data/politics_comments.csv')
```

```python
co
```

```python
co['rank'] = co.groupby('article_id')['timestamp'].rank(method='dense').astype(int)
co = co[co['rank'] <= 10] # select only first 10 comments
co['total_upvotes'] = co.groupby('article_id')['upvotes'].transform('sum')
co['total_comments'] = co.groupby('article_id')['upvotes'].transform('count')
```

```python
co
```

```python
co = co[co['total_upvotes'] > 20] # do not consider articles with under 10 upvotes
```

```python
co
```

```python
co = co[co['total_comments'] == 10] # remove articles with over under 10 comments. 
```

```python
co
```

```python
co['rel_upvotes'] = co.apply(lambda row: (row.upvotes / row.total_upvotes) * 100, axis=1)
```

```python
# filterd = co[co['comment_text'].str.len() > 50]
filterd = co # don't filter for now
```

```python
filterd
```

```python
# the perc has to be under 1.0! Because we take from the bot and top!
top_bot_perc = 0.25
num_ranks = 10

num_rows=filterd.shape[0]
N = (num_rows * top_bot_perc) / num_ranks
N = int(N)
print(N)
print(N * num_ranks * 2)
print(num_rows)
```

```python
res_pos = filterd.groupby("rank", group_keys=False).apply(lambda g: g.nlargest(N, "rel_upvotes", keep="last"))
```

```python
res_pos
```

```python
res_neg = filterd.groupby("rank", group_keys=False).apply(lambda g: g.nsmallest(N, "rel_upvotes", keep="first"))
```

```python
res_neg
```

```python
res_pos['class'] = 1
res_neg['class'] = 0
```

```python
if (top_bot_perc == 0.5):
# There is a problem when the want to bin the *whole* dataset. It would result in 2 duplicates. Most likely due to 
# rel_upvotes is the same for multiple values and it's not possible to have a clear cut.
    res = pd.merge(filterd, res_pos, how='left')
    res['class'] = res['class'].fillna(0)
else:
    res = pd.concat([res_pos, res_neg])
```

```python
assert(res[res['class'] == 0].shape[0] == res[res['class'] == 1].shape[0]) # ensure same number of classes
```

```python
assert(res['comment_id'].is_unique) # make sure we don't have duplicates
```

```python
res.to_csv("/mnt/data/group07/johannes/proc_data/classes_politics_comments_" + str(top_bot_perc) + "_fixed.csv")
```
