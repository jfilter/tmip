```python
import pandas as pd
import math
import numpy as np
```

```python
co = pd.read_csv('/data/tmip/only_pol/comments.csv')
```

```python
# clean data
# co = co[co['comment_text'].notnull()]
# co = co[co['comment_text'].str.len() > 100]
# co.to_csv('/data/tmip/only_pol/comments_cleaned_100.csv')
# co = pd.read_csv('/data/tmip/only_pol/comments_cleaned_100.csv')
```

```python
co['rank'] = co.groupby('article_id')['timestamp'].rank(method='dense').astype(int)
```

```python
co[co['article_id'] == 168]
```

```python
co['total_upvotes'] = co.groupby('article_id')['upvotes'].transform('sum')
co['total_comments'] = co.groupby('article_id')['upvotes'].transform('count')
```

```python
# co = co[co['total_comments'] > 20] # do not consider articles with under 10 upvotes
```

```python
co = co[co['total_upvotes'] > 10] # do not consider articles with under 10 upvotes
```

```python
co = co[co['total_comments'] > 50] # remove articles with over 50 comments. 
```

```python
# this was used before to calc some random score
# co['score'] = co.apply(lambda row: (row.upvotes / math.log(row.total_upvotes + 1)) + (row['rank'] / row.total_comments), axis=1)
# co['bin'] = pd.qcut(co['score'], q=3, labels=range(3))
# co.to_csv('/data/tmip/only_pol/comments_cleaned_100_3classes.csv')
```

```python
co['rel_upvotes'] = co.apply(lambda row: (row.upvotes / row.total_upvotes) * 100, axis=1)
```

```python
co
```

```python
r = co.groupby('rank').median().reset_index()
```

```python
r = r[['rank', 'rel_upvotes']].rename(columns={'rel_upvotes': 'med_rel_upvotes'})
```

```python
r
```

```python
# This was used before to binary classify into below / above the median
# res['bin'] = np.where(res['med_rel_upvotes'] > res['rel_upvotes'], 0, 1)
```

```python
res = co.join(r, how='left', on='rank', rsuffix='r_')
res = res[res['rank'] <= 30]
res['med_rel_upvotes_diff'] = (res['med_rel_upvotes'] - res['rel_upvotes']) ** 2
res['class'] = pd.qcut(res['med_rel_upvotes_diff'], q=10, labels=range(10))
res = res[(res['class'] == 0) | (res['class'] == 9)]
res['bin'] = np.where(res['class'] > 0, 1, 0)
```
