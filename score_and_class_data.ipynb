# Apply extracted features and filter

Uses
- /mnt/data/group07/johannes/proc_data/politics_comments.csv

For generating
- /mnt/data/group07/johannes/proc_data/classes_politics_comments_{top_bot_perc}.csv [top_bot_perc=1.0]



```python
import pandas as pd
import math
import numpy as np
```

```python
co = pd.read_csv('/mnt/data/group07/johannes/proc_data/politics_comments.csv')
```

```python
co
```

```python
co['rank'] = co.groupby('article_id')['timestamp'].rank(method='dense').astype(int)
co = co[co['rank'] <= 10] # select only first 10 comments
co['total_upvotes'] = co.groupby('article_id')['upvotes'].transform('sum')
co['total_comments'] = co.groupby('article_id')['upvotes'].transform('count')
```

```python
co
```

```python
co = co[co['total_upvotes'] > 20] # do not consider articles with under 10 upvotes
```

```python
co
```

```python
co = co[co['total_comments'] == 10] # remove articles with over under 10 comments. 
```

```python
co
```

```python
co['rel_upvotes'] = co.apply(lambda row: (row.upvotes / row.total_upvotes) * 100, axis=1)
```

```python
# filterd = co[co['comment_text'].str.len() > 50]
filterd = co # don't filter for now
```

```python
filterd
```

```python
top_bot_perc = 1.0
num_ranks = 10

num_rows=filterd.shape[0]
N = (num_rows * top_bot_perc) / num_ranks
N = int(N)
N
```

```python
res_pos = filterd.groupby("rank", group_keys=False).apply(lambda g: g.nlargest(N, "rel_upvotes"))
```

```python
res_pos
```

```python
res_neg = filterd.groupby("rank", group_keys=False).apply(lambda g: g.nsmallest(N, "rel_upvotes"))
```

```python
res_neg
```

```python
res_pos['class'] = 1
res_neg['class'] = 0
```

```python
res_pos
```

```python
res = pd.concat([res_pos, res_neg])
res.to_csv("/mnt/data/group07/johannes/proc_data/classes_politics_comments_" + str(top_bot_perc) + ".csv")
```
